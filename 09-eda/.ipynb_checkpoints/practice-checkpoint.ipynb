```python
# importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load iris data set
iris = pd.read_csv("iris.csv")
```

```python
# (Q1) number of data points and features?
iris.shape
```

```python
# (Q2) column names in dataset?
iris.columns
```

```python
# (Q3) how many data points for a class?
iris["species"].value_counts()
# given dataset is balanced
```

```python
# make a 2D plot
plt.plot(iris["sepal_length"], iris["sepal_width"], "bo")
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
plt.grid()
plt.show()
```

```python
# make a plot with different colored points corresponding to each class
sns.set_style("whitegrid")
sns.FacetGrid(iris, hue="species", height=4) \
   .map(plt.scatter, "sepal_length", "sepal_width") \
   .add_legend();
plt.show()
# observations-
# setosa can easily be separated from versicolor and virginica by drawing a line (linearly separable)
# but versicolor and virginica have significant overlap
```

```python
# pair plots
sns.pairplot(iris, hue="species", height=3)
plt.show()
```

```python
# histograms
sns.FacetGrid(iris, hue="species", height=5) \
   .map(sns.distplot, "petal_length") \
   .add_legend()
plt.show();
```

```python
sns.FacetGrid(iris, hue="species", height=5) \
   .map(sns.distplot, "petal_width") \
   .add_legend()
plt.plot()
```

```python
sns.FacetGrid(iris, hue="species", height=5).map(sns.distplot, "sepal_length").add_legend()
plt.plot()
```

```python
sns.FacetGrid(iris, hue="species", height=5).map(sns.distplot, "sepal_width").add_legend()
plt.plot()
```

```python
# mean
print(np.mean(iris["petal_length"][iris["species"] == "setosa"]))
print(np.mean(iris["petal_length"][iris["species"] == "versicolor"]))
print(np.mean(iris["petal_length"][iris["species"] == "virginica"]))
print()
# standard deviation
print(np.std(iris["petal_length"][iris["species"] == "setosa"]))
print(np.std(iris["petal_length"][iris["species"] == "versicolor"]))
print(np.std(iris["petal_length"][iris["species"] == "virginica"]))
print()
# median
print(np.median(iris["petal_length"][iris["species"] == "setosa"]))
print(np.median(iris["petal_length"][iris["species"] == "versicolor"]))
print(np.median(iris["petal_length"][iris["species"] == "virginica"]))
```

```python
# quantiles
print(np.percentile(iris["petal_length"][iris["species"] == "setosa"], np.arange(25,100,25)))
print(np.percentile(iris["petal_length"][iris["species"] == "versicolor"], np.arange(25,100,25)))
print(np.percentile(iris["petal_length"][iris["species"] == "virginica"], np.arange(25,100,25)))
print()
# 90th percentile
print(np.percentile(iris["petal_length"][iris["species"] == "setosa"], 90))
print(np.percentile(iris["petal_length"][iris["species"] == "versicolor"], 90))
print(np.percentile(iris["petal_length"][iris["species"] == "virginica"], 90))
```

```python
# median absolute deviation
from statsmodels import robust
print(robust.mad(iris["petal_length"][iris["species"] == "setosa"]))
print(robust.mad(iris["petal_length"][iris["species"] == "versicolor"]))
print(robust.mad(iris["petal_length"][iris["species"] == "virginica"]))
```

```python
# box plots with whiskers
# whiskers = 1.5*IQR
# uses percentiles
# gies 25,50,75 percentile values i.e quantiles
sns.boxplot(x="species", y="petal_length",  data=iris)
plt.plot();
```

```python
# violin plots
# gives both pdf and box plots
sns.violinplot(x="species", y="petal_length", data=iris)
plt.plot();
```

```python
# 2d density plots
sns.jointplot(x="petal_length", y="petal_width", data=iris[iris["species"] == "setosa"], kind="kde", fill=True)
plt.plot();
```

```python
# exercise
# haberman dataset
# load data
haberman = pd.read_csv("haberman.csv")
```

```python
# giving logical column names
haberman.columns = ["Age", "Op_Year", "axil_nodes", "Surv_status"]
```

```python
# number of data points and number of features
haberman.shape
# data points = 305
# features = 3
# label = 1
```

```python
# split classes
haberman_1 = haberman[haberman["Surv_status"] == 1]
haberman_2 = haberman[haberman["Surv_status"] == 2]
```

```python
# number of data points of every class
haberman["Surv_status"].value_counts()
# dataset is not balanced
```

```python
# main objective- given the age, operation year, and axillary nodes determine the survival status of the patient
```

```python
# univariate analysis
# box plots
sns.boxplot(x="Surv_status", y="Age", data=haberman)
plt.plot();
# gives qunatiles of age for feature type 1 and feature type 2
```

```python
sns.boxplot(x="Surv_status", y="Op_Year", data=haberman)
plt.plot();
# gives quantiles of opearation year for feature type 1 and type 2
```

```python
sns.boxplot(x="Surv_status", y="axil_nodes", data=haberman)
plt.plot();
# gives quantiles of axil_nodes for feature type 1 and type 2
```

```python
# quantiles
for feature in haberman.columns[0:3]:
    print(f"Quantile for {feature} is-")
    print(np.percentile(haberman[feature], np.arange(25,100,25)))
    print()
```

```python
# violin plots
sns.violinplot(x="Surv_status", y="Age", data=haberman)
plt.plot();
# gives box plot as well as pdf
```

```python
sns.violinplot(x="Surv_status", y="Op_Year", data=haberman)
plt.plot();
```

```python
sns.violinplot(x="Surv_status", y="axil_nodes", data=haberman)
plt.plot();
```

```python
# bivariate anaylysis
# pairplots
sns.pairplot(data=haberman, hue="Surv_status", kind="scatter", height=4, palette=["black", "red"])
plt.plot();
# features are not linearly separable as seen from the pair plots
```

```python
# multivariate analysis
# contour plots
sns.jointplot(x="Age", y="Op_Year", data=haberman, kind="kde", fill=True)
plt.plot();
```

```python
sns.jointplot(x="axil_nodes", y="Age", data=haberman, kind="kde", fill=True)
plt.plot();
```
